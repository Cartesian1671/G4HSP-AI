```python
def prepare_corpus(doc_clean):
    """
    This function prepares the corpus by creating a dictionary and document-term matrix.
    Input: 
    - doc_clean: A list of tokenized documents
    Return: 
    - dictionary: A mapping from word IDs to words
    - doc_term_matrix: A list of bag-of-word vectors representing each document
    """
    # Creating the term dictionary of our corpus, where every unique term is assigned an index. 
    dictionary = gensim.corpora.Dictionary(doc_clean)

    # Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
    doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
    
    return dictionary, doc_term_matrix
```

```python
import requests
import re
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from gensim import corpora
from gensim.models import LsiModel, LdaModel
from nltk.stem import WordNetLemmatizer
from gensim.models.coherencemodel import CoherenceModel

# Load the articles as a list
with open('/content/CollatedText.txt', 'r', encoding="utf-8") as a:
    docs = [doc.strip() for doc in a.readlines()]

def preprocess_data(doc_set):
    tokenizer = RegexpTokenizer(r'\w+')
    en_stop = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()   # Use NLTK's WordNetLemmatizer
    texts = []

    for i in doc_set:
        raw = i.lower()
        # remove references like [3,4] and (Fig. 1)
        raw = re.sub(r'\[[^\]]*\]', '', raw)
        raw = re.sub(r'\(Fig\. .*?\)', '', raw)
        # split into tokens
        tokens = tokenizer.tokenize(raw)
        # remove stopwords
        stopped_tokens = [i for i in tokens if not i in en_stop]
        # lemmatize tokens
        lemmatized_tokens = [lemmatizer.lemmatize(i) for i in stopped_tokens]
        texts.append(lemmatized_tokens)

    return texts

def prepare_corpus(doc_clean):
    dictionary = corpora.Dictionary(doc_clean)
    doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
    return dictionary, doc_term_matrix

clean_docs = preprocess_data(docs)
dictionary, doc_term_matrix = prepare_corpus(clean_docs)

# LSI Model
lsi_model = LsiModel(doc_term_matrix, num_topics=20, id2word = dictionary)
print("LSI Model:")
for idx, topic in lsi_model.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))

# LDA Model
lda_model = LdaModel(doc_term_matrix, num_topics=20, id2word = dictionary)
print("LDA Model:")
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))
```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, doc_term_matrix, dictionary)
pyLDAvis.display(vis)
```
